## Groupby

```python
# Create a dataframe using dictionary of your choice

# data =

data=pd.DataFrame({'College':['IIM','IIT','NIT','BITS'],'Exams':['CAT','JEE-ADV','JEE','BITSAT'],'Percentile':[99.9,99,95,90]})
data
```

```python
# Use Groupby of single column with aggregate sum()

data.groupby('College')['Exams'].sum()
```

```python
# Use Groupby of single column with aggregate count()

data.groupby('College')['Exams'].count()
```

```python
# Use Groupby of single column with aggregate min() and max()

print(data.groupby('College')['Percentile'].min())
print(data.groupby('College')['Percentile'].max())
```

```python
# Use Groupby of any 2 columns with aggregate mean()

data.groupby(['College','Exams'])['Percentile'].mean()
```

```python
# Use Groupby of any 2 columns with aggregate min() and max()

data.groupby(['College','Exams'])['Percentile'].min()
data.groupby(['College','Exams'])['Percentile'].max()
```
